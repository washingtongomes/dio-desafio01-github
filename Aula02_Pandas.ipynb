```python
import pandas as pd

```

```python
df1 = pd.read_excel("/content/drive/MyDrive/dio-datasets/Aracaju.xlsx")
df2 = pd.read_excel("/content/drive/MyDrive/dio-datasets/Fortaleza.xlsx")
df3 = pd.read_excel("/content/drive/MyDrive/dio-datasets/Natal.xlsx")
df4 = pd.read_excel("/content/drive/MyDrive/dio-datasets/Recife.xlsx")
df5 = pd.read_excel("/content/drive/MyDrive/dio-datasets/Salvador.xlsx")
```

```python
df5.head()
```

```python
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
df.head()
```

```python
df.tail()
```

```python
#amostra
df.sample(5)
```

```python
#tipos
df.dtypes
```

```python
df["LojaID"] = df["LojaID"].astype("object")
```

```python
df.dtypes
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#subst. valores nulos pela media
df["Vendas"].fillna(df["Vendas"].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.sample(15)
```

```python
df["Vendas"].mean()
```

```python
#Trabalhando com datas
df["Data"] = df["Data"].astype("int64")
```

```python
df.dtypes
```

```python

df["Data"] = pd.to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
#agrupamento por ano
df.groupby(df["Data"].dt.year)["Vendas"].sum()
```

```python
#criar coluna
df["Ano_Venda"] = df["Data"].dt.year
```

```python
df.sample(5)
```

```python
df["Mes_Venda"], df["Dia_Venda"] = df["Data"].dt.month, df["Data"].dt.day
```

```python
df.sample(5)
```

```python
#data mais antiga
df["Data"].min()
```

```python
#calculando diferenca dias
df["diferenca_dias"] = df["Data"] - df["Data"].min()
```

```python
df.sample(5)
```

```python
#filtrando vendas 2019 e por mes 3
vendas_mes_marco_2019 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)] 
```

```python
vendas_mes_marco_2019.sample(20)
```
